```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("netflix_titles.csv")
print("The shape is: ", df.shape)
print("Here is the quick view::::")
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
print("There are "+str(df.isnull().sum().sum())+" missing values")
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title('Null Values Heatmap')
plt.show()
```

```python
df.fillna('', inplace=True)
```

```python
df.isnull().sum()
```

```python
df_movie = df[df['type']=='Movie'].copy()
df_movie.head()
```

```python
df_shows =df[df['type']=='TV Show'].copy()
df_shows.head()
```

```python
plt.figure(figsize=(7,5))
g = sns.countplot(df.type)
plt.title("Count of Movies and TV Shows")
plt.xlabel("Type (Movie/TV Show)")
plt.ylabel("Total Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("% of Netflix Titles that are either Movies or TV Shows")
g = plt.pie(df.type.value_counts(), explode=(0.025,0.025), labels=df.type.value_counts().index,autopct='%1.1f%%', startangle=180);
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,7))
g = sns.countplot(df.rating, hue=df.type)
plt.title("Ratings for Movies & TV Shows")
plt.xlabel("Rating")
plt.ylabel("Total count")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(19, 5))
g1 = sns.countplot(df.rating, ax=ax[0]);
g1.set_title("Ratings for Movies")
g1.set_xlabel("Rating")
g1.set_ylabel("Total Count")
g2 = sns.countplot(df.rating, ax=ax[1]);
g2.set(yticks=np.arange(0,1600,200))
g2.set_title("Ratings for TV Shows")
g2.set_xlabel("Rating")
g2.set_ylabel("Total Count")
fig.show()
```

```python
df['year'] = pd.DatetimeIndex(df['date_added']).year
df_movie['year'] = pd.DatetimeIndex(df_movie['date_added']).year
df_shows['year'] = pd.DatetimeIndex(df_shows['date_added']).year
df['month'] = pd.DatetimeIndex(df['date_added']).month
df_movie['month'] = pd.DatetimeIndex(df_movie['date_added']).month
df_shows['month'] = pd.DatetimeIndex(df_shows['date_added']).month
```

```python
df.head()
```

```python
df_movie.head()
```

```python
df_shows.head()
```

```python
movie_year = df_movie['year'].value_counts().to_frame().reset_index().rename(columns={'index': 'year','year':'count'})
show_year = df_shows['year'].value_counts().to_frame().reset_index().rename(columns={'index': 'year','year':'count'})
netflix_year = df['year'].value_counts().to_frame().reset_index().rename(columns={'index': 'year','year':'count'})

netflix_year
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=movie_year, x='year', y='count')
sns.lineplot(data=show_year, x='year', y='count')
sns.lineplot(data=netflix_year, x='year', y='count')
ax.set_xticks(np.arange(2008, 2020, 1))
plt.title("Total content added each year (up to 2019)")
plt.legend(['Total','Movie','TV Show'])
plt.ylabel("Releases")
plt.xlabel("Year")
plt.show()
```

```python
filtered_countries = df.set_index('title').country.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);


plt.figure(figsize=(7,9))
g = sns.countplot(y = filtered_countries, order=filtered_countries.value_counts().index[:20])
plt.title('Top 20 Countries on Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()
```

```python
filtered_genres = df.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);

plt.figure(figsize=(7,9))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:20])
plt.title('Top 20 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```

```python
filtered_directors = df[df.director != ''].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:10], palette='mako')
plt.show()
```

```python
filtered_cast = df[df.cast != ''].set_index('title').cast.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_cast, order=filtered_cast.value_counts().index[:10], palette='rocket')
plt.show()
```
